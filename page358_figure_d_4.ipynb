```python
# A 爬升率约束

import matplotlib.pyplot as plt
import numpy as np

beta = 0.95; # 查教材图3.12
sigma = 1.056/1.225; # 1.056为5000 ft高度空气密度
Ma = 250 * 0.51444 / 340.0 # kn转换为m/s
alpha = (0.568 + 0.25*(1.2-Ma)**3)*sigma**0.6
A = 7.50; # 查教材表3.4
e = 1.78*(1.0-0.045*A**0.68) - 0.64; #公式3.36，注意此式适用于直机翼，后掠翼飞机适用公式3.37
K1 = 1./np.math.pi/A/e
g = 9.8; #常数
rho = 1. ## 1.056为5000 ft高度空气密度
V = 250 * 0.51444 # kn转换为m/s
q = 1/2*rho*V**2
K2 = 0
CD0 = 0.02
dh_dt = 36.0 #爬升率 m/s

m0_S = np.linspace(200, 500, 100)
F0_m0g = beta / alpha * (K1 *beta*g/q + K2 + CD0/(beta*g/q * m0_S) + 1/V * dh_dt) #教材公式3.23

# Create the plot
plt.plot(m0_S, F0_m0g)

# Show the plot
plt.show()
```

```python
# B 爬升率约束

import matplotlib.pyplot as plt
import numpy as np

beta = 0.8; # 查教材图3.12
sigma = 0.905/1.225; # 0.905为10000 ft高度空气密度
Ma = 0.63 # kn转换为m/s
alpha = (0.568 + 0.25*(1.2-Ma)**3)*sigma**0.6
A = 7.50; # 查教材表3.4
e = 1.78*(1.0-0.045*A**0.68) - 0.64; #公式3.36，注意此式适用于直机翼，后掠翼飞机适用公式3.37
K1 = 1./np.math.pi/A/e
g = 9.8; #常数
rho = 1. ## 1.056为5000 ft高度空气密度
V = Ma * 340.0 # Ma转换为m/s
q = 1/2*rho*V**2
K2 = 0
CD0 = 0.02
dh_dt = 15.0 #爬升率 m/s

m0_S = np.linspace(200, 500, 100)
F0_m0g = beta / alpha * (K1 *beta*g/q + K2 + CD0/(beta*g/q * m0_S) + 1/V * dh_dt) #教材公式3.23

# Create the plot
plt.plot(m0_S, F0_m0g)

# Show the plot
plt.show()
```

```python
# C 着陆距离约束

import matplotlib.pyplot as plt

x_LGR = 1000.0; # m
rho = 1.225 # 查海平面大气数据 kg/m^3
mu = 0.2; # 摩擦系数
C_Lmax = 2.2; # 查教材表3.3
k_TD = 1.3; # 安全系数
beta = 0.7; # 查教材图3.12

m0_S = x_LGR * rho * C_Lmax * mu / k_TD**2 / beta #教材公式3.34

# Create the plot
plt.axvline(x = m0_S)

# Show the plot
plt.show()
```

```python
# D 起飞距离约束

import matplotlib.pyplot as plt

x_LGR = 800.0; # m
rho = 1.225 # 查海平面大气数据 kg/m^3
C_Lmax = 1.8; # 查教材表3.3
k_LOF = 1.3; # 安全系数
beta = 1.0; # 查教材图3.12
alpha = 1.0

m0_S = np.linspace(200, 500, 100)
F0_m0g = beta**2/alpha * k_LOF**2/x_LGR/rho/C_Lmax * m0_S #教材公式3.22

# Create the plot
plt.plot(m0_S, F0_m0g)

# Show the plot
plt.show()
```
